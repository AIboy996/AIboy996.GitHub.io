### 画图

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y, label="$\sin(x) \delta$")
plt.plot(x, y2, label="cos(x)")
plt.legend()
plt.title("Sine and Cosine Functions")
````

### 表格

````python
import pandas as pd

df = pd.DataFrame({'x': x, 'y': y, 'y2': y2})

df
````
